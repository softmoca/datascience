# Something before we start Python
- In the **assignment** statement, **the expression on the right is evaluated first. That value is then assigned to the variable on the left.**
  - When assigning a variable in Python, you are creating a reference to the object (i.e pointer) , 메모리의 어딘가에 그 object를 만들어 저장하고 (그릇) 그 주소를 알려줌 (그릇의 장소).
  - **If you assign the variable to a new variable, you are only assigning the reference**, not creating a new object. (try the code below)
- Assignment in Python means one thing, and one thing only: The variable named on the left should now refer to the value on the right. In other words, when I said: y = x. Python doesn't read this as, “y should now refer to the variable x.” Rather, it read it as, **“y should now refer to whatever value x refers to.”**


```python
# try to see how Python works on variables
# id(object):  the identity of an object. The identity is a unique integer 
# for that object during its lifetime. This is also the address of the 
# object in memory.

a = [1, 2, 3]
b = a
b.append(4)
print(a, b)
print(id(a), id(b))   # the same object
```

```python
a = 10
b = a
b += 20

print(a, b)
print(id(a), id(b))   # immutable, so created a new object (different objects)
```

```python
# mutation and iteration - be careful !
# do not mutate a list as you iterating over it !
def remove_dups(L1, L2):   # wrong code
    for e in L1:
        if e in L2:
            L1.remove(e)
L1 = [1,2,3,4]
L2 = [1,2,5,6]
remove_dups(L1, L2)   # L1 is [2,3,4] not [3,4], why?
print(L1)
```

- L1 is [2,3,4] not [3,4] Why?
  - Python uses an internal counter to keep track of index it is in the loop
  - mutating changes the list length but Python doesn’t update the counter
  - loop never sees element 2

```python
# correct one
def remove_dups(L1, L2):
    L1_copy = L1[:]   # clone list first
                      # note that L1_copy = L1 does not clone
    for e in L1_copy:
        if e in L2:
            L1.remove(e)
L1 = [1,2,3,4]
L2 = [1,2,5,6]
remove_dups(L1, L2)
print(L1)
```

- Functions and methods are called callable as they can be called.
- In fact, any object which implements the special __call__() method is termed callable.

```python
def make_pretty(func):
    def inner():
        print("I got decorated")
        func()
    return inner

def ordinary():
    print("I am ordinary")
```

```python
ordinary()

# let's decorate this ordinary function
pretty = make_pretty(ordinary)
pretty()
```

```python
@make_pretty
def ordinary1():
    print("I am ordinary")

# is equivalent to
def ordinary2():
    print("I am ordinary")
ordinary = make_pretty(ordinary)
```

```python
ordinary1()   # decorated
```

```python
ordinary2()
```

```python
# one more example
# decorating functions with parameters
def divide1(a, b):
    return a/b
divide1(2,5), divide1(2,0)
```

- Now, we want to make a decorator to check for this case that will cause the error.

```python
def smart_divide(func):
    def inner(a, b):
        print("I am going to divide", a, "and", b)
        if b == 0:
            print("Whoops! cannot divide")
            return

        return func(a, b)
    return inner

@smart_divide
def divide2(a, b):
    print(a/b)
```

```python
divide2(2,5)
```

```python
divide2(2,0)
```

```python
# functions can be decorated multiple times 
def star(func):
    def inner(*args, **kwargs):
        print("*" * 30)
        func(*args, **kwargs)
        print("*" * 30)
    return inner

def percent(func):
    def inner(*args, **kwargs):
        print("%" * 30)
        func(*args, **kwargs)
        print("%" * 30)
    return inner

def printer1(msg):
    print(msg)

@star
@percent
def printer2(msg):
    print(msg)

printer1("Hello")
```

```python
printer2("Hello")
```

```python
# is equivalent to  (order matters)
def printer3(msg):
    print(msg)
printer = star(percent(printer3))
printer("Hello")
```

- one more example (한국말 설명)
  - Wrapper나 Wrapper Function을 이용해 어떤 함수들의 실행 이전이나 이후에 공통으로 선행해야하는 기능들이 실행되게 하는 기법
  - 클래스나 함수로 Decorator 설정가능

```python
# one more example
def decorator_function(original_function):
    def wrapper_function():
        print('{} 함수가 호출되기전 입니다.'.format(original_function.__name__))
        return original_function()
    return wrapper_function

@decorator_function
def display_1():
    print('display_1 함수가 실행됐습니다.')

@decorator_function
def display_2():
    print('display_2 함수가 실행됐습니다.')
    
display_1()
display_2()
```

```python
# another example (following two examples are the same !)
@tf.function
def callable_func(a, b):
  return tf.matmul(a, b)

a1 = tf.constant([[1.0, 2.0]])
b1 = tf.constant([[2.0], [3.0]])
callable_func(a1, b1).numpy()
```

```python
# same as the above
import tensorflow as tf
def callable_func(a, b):
  return tf.matmul(a, b)

# Create a `Function` object that contains a graph
function_that_uses_graph = tf.function(callable_func)

# Make some tensors
a1 = tf.constant([[1.0, 2.0]])
b1 = tf.constant([[2.0], [3.0]])

# It works!
function_that_uses_graph(a1, b1).numpy()
```

```python
# let's compare execution time (Graph version is faster!)
# but, not always...

import timeit
class SequentialModel(tf.keras.Model):
  def __init__(self, **kwargs):
    super().__init__(**kwargs)
    self.flatten = tf.keras.layers.Flatten(input_shape=(28, 28))
    self.dense_1 = tf.keras.layers.Dense(128, activation="relu")
    self.dropout = tf.keras.layers.Dropout(0.2)
    self.dense_2 = tf.keras.layers.Dense(10)

  def call(self, x):
    x = self.flatten(x)
    x = self.dense_1(x)
    x = self.dropout(x)
    x = self.dense_2(x)
    return x

input_data = tf.random.uniform([60, 28, 28])

eager_model = SequentialModel()
graph_model = tf.function(eager_model)

print("Eager time:", timeit.timeit(lambda: eager_model(input_data), number=10000))
print("Graph time:", timeit.timeit(lambda: graph_model(input_data), number=10000))
```

### Multiple inheritance

```python
# multiple inheritance
class Family:
    familyword = ""
    def show_family(self):
        print("This is our family:")
 
# Father class inherited from Family
class Father(Family):
    fathername = ""
    def show_father(self):
        print(self.fathername)
 
# Mother class inherited from Family
class Mother(Family):
    mothername = ""
    def show_mother(self):
        print(self.mothername)
 
# Son class inherited from Father and Mother classes
class Son(Father, Mother):
    sonname = ""
    def show_parent(self):
        print("Father :", self.fathername)
        print("Mother :", self.mothername)
        print("Son :", self.sonname)
 
s1 = Son()  # Object of Son class
s1.familyword = "truth"
s1.fathername = "Mark"
s1.mothername = "Sonia"
s1.sonname = "Michael"
```

```python
s1.familyword, s1.fathername, s1.mothername, s1.sonname
```

```python
s1.show_family(), s1.show_father(), s1.show_mother(), s1.show_parent()
```

# Exeptions and Assertions
- try and except: exceptions raised by any statement in body of *try* are
handled by the *except* statement and execution continues
with the body of the *except* statement
- else: body of this is executed when execution of associated
try body **completes with no exceptions**
- finally: body of this is **always executed** after try, else and
except clauses, even if they raised another error or
executed a break, continue or return

- assert: want to be sure that assumptions on state of
computation are as expected

```python
try:
    a = int(input("Tell me one number:"))
    b = int(input("Tell me another number:"))
    print(a/b)
except:
    print("Bug in user input.")
else:
    print("no error.")
finally:
    print("anyway, it's done.")
```

# Python coding Example (1) : Month Carlo

```python
# first version of calculating pi
import numpy as np

# simple code for calculating pi
N=100
x=np.random.rand(N).round(2)          # simple version
y=np.random.rand(N).round(2)

cnt=0
for n in range (N):
    if ((x[n]*x[n]) + (y[n]*y[n]) < 1.0) :
        cnt += 1
    else :
        continue
pi = 4*cnt/N
print(pi)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(4,4))      # plotting a circle
plt.scatter(x,y, s=1)          # s: marker size

a = np.linspace(0,1,100)
b = (1-a*a)**0.5

plt.plot(a,b, c='r')
```

# Python coding Example (2) : Squre Root of x
- simple guess and check algorithm
- Newton-Raphson method (xi+1 = xi - f(xi)/f'(xi))
  - start with a guess, g
  - if g*g is close enough to x, stop (g is the answer)
  - otherwise, make a new guess by averaging g and x/g
  - using the new guess, repeat process

```python
# Guess anf Check algotithm (only works for x >= 1) - poor algorithm
import numpy as np
epsilon = 0.01

def gc_sqrt(x):
    if x < 0.0:
        return (np.nan)
    iter = 0
    left, right = 0, x
    guess = (right + left)/2.0
    while abs (guess**2 - x) > epsilon:
        if guess **2 < x:
            left = guess
        else:
            right = guess
        guess = (right + left)/2.0
        iter += 1

    return guess, iter
```

```python
gc_sqrt (77777), gc_sqrt (77777)[0]**2
```

```python
# Newton-Raphson algorithm
epsilon = 0.01
max_iter = 100

def gc_sqrt2(x):
    g = np.random.rand()
    i = 0
    while i < max_iter:
        diff = g**2 - x
        if abs(diff) < epsilon:
            return g, i
        else:
            g = g - diff / (2.0 * g)
            # g = 0.5 * (g + x/g)
        i += 1
    
    return g, i
```

```python
# example 2
import numpy as np
import matplotlib.pyplot as plt

# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2, 1, 1)

# Make the first plot
plt.plot(x, y_sin)
plt.title('Sine')
plt.subplots_adjust(wspace=1, hspace=0.5)

# Set the second subplot as active, and make the second plot.
plt.subplot(2, 1, 2)
plt.plot(x, y_cos, c='r')
plt.title('Cosine')

# Show the figure.
plt.show()
```

# Images
- You can use the 'imshow' function to show images
- image data format: png의 경우는 무손실 그림 저장 포맷이고, RGBA(Red, Green, Blue, Alpha)라는 네가지 레이어로 이미지를 저장합니다. jpg의 경우는 손실 그림 저장 포맷이고, RGB(Red, Green, Blue)로 이미지를 저장합니다.

```python
# to show image 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('cat161.jpg')
img_tinted = img * [0.90, 0.90, 0.70]

# Show the original image
plt.subplot(1, 2, 1)
plt.imshow(img)

# Show the tinted image
plt.subplot(1, 2, 2)

# A slight gotcha with imshow is that it might give strange results
# if presented with data that is not uint8. To work around this, we
# explicitly cast the image to uint8 before displaying it.
plt.imshow(np.uint8(img_tinted))
plt.show()
```
